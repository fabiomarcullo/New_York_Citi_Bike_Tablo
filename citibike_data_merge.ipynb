```python
import os
import pandas as pd

# Read the path
file_path = "citibike_data"

# List all the files from the directory
file_list = os.listdir(file_path)
file_list

df_append = pd.DataFrame()
```

```python
# Append all files together
for file in file_list:
    file_abs_path = os.path.join(file_path, file)
    df_temp = pd.read_csv(file_abs_path)
    df_append = df_append.append(df_temp, ignore_index=True)

df_append
```

```python
# Save the appended DataFrame as a CSV file
output_file = "citibike_data_merge.csv"
df_append.to_csv(output_file, index=False)

print("Data appended and saved as", output_file)
```
